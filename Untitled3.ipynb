```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 1. Data load karein (encoding ke saath taaki error na aaye)
df = pd.read_csv('stores_sales_forecasting.csv', encoding='latin1')

# 2. Data saaf karein (Date aur Sales column ka use)
# Agar aapke dataset mein column ka naam alag hai toh yahan badal dein
df['Date'] = pd.to_datetime(df['Date'])
monthly_data = df.groupby('Date')['Sales'].sum().reset_index()
monthly_data['Time_Index'] = np.arange(len(monthly_data))

# 3. Linear Regression Model train karein
X = monthly_data[['Time_Index']]
y = monthly_data['Sales']

model = LinearRegression()
model.fit(X, y)

# 4. Predictions nikaalein
predictions = model.predict(X)

# 5. Graph visualize karein
plt.figure(figsize=(12, 6))
plt.plot(monthly_data['Date'], y, label='Actual Sales (Data)', color='blue', marker='o')
plt.plot(monthly_data['Date'], predictions, label='Predicted Trend (Model)', color='red', linestyle='--')
plt.title('Store Sales Forecasting Trend')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()

# 6. Evaluation Score
print(f"Model Score: {model.score(X, y):.2f}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 1. Data load karein
df = pd.read_csv('stores_sales_forecasting.csv', encoding='latin1')

# 2. Sahi column names ka use karke process karein
# Aapke dataset ke hisaab se column name 'Order Date' hai
df['Order Date'] = pd.to_datetime(df['Order Date'])
monthly_data = df.groupby('Order Date')['Sales'].sum().reset_index()
monthly_data['Time_Index'] = np.arange(len(monthly_data))

# 3. Model Training
X = monthly_data[['Time_Index']]
y = monthly_data['Sales']
model = LinearRegression()
model.fit(X, y)

# 4. Predictions
predictions = model.predict(X)

# 5. Graph
plt.figure(figsize=(12, 6))
plt.plot(monthly_data['Order Date'], y, label='Actual Sales', color='blue')
plt.plot(monthly_data['Order Date'], predictions, label='Predicted Trend', linestyle='--', color='red')
plt.title('Store Sales Predictive Analytics')
plt.xlabel('Order Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```
